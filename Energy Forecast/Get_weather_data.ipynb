```python
import requests

def get_population(city, country):
    base_url = "http://api.geonames.org/searchJSON"
    username = "YOUR_GEONAMES_USERNAME"

    params = {
        "q": city,
        "country": country,
        "maxRows": 1,
        "username": username
    }

    response = requests.get(base_url, params=params)
    data = response.json()
    
    if "geonames" in data and len(data["geonames"]) > 0:
        population = data["geonames"][0].get("population")
        return population

    return None

top_cities = ["Toronto", "Ottawa", "Mississauga", "Brampton", "Hamilton"]
country = "CA"  # Country code for Canada

population_data = {}

for city in top_cities:
    population = get_population(city, country)
    population_data[city] = population

print("Population of the top five cities in Ontario, Canada:")
for city, population in population_data.items():
    print(f"{city}: {population}")
```

```python
Toronto_pop = (3025647)
Mississanga_pop =  (771891)
Ottawa_pop = (1071868)
Brampton_pop = (745557)
Hamilton_pop = (597010)

Population = [Toronto_pop,Mississanga_pop,Ottawa_pop,Brampton_pop,Hamilton_pop]

#Population_weights = []
Total_pop = sum(Population)
```

```python
def population_weight(Population, Total_pop):
    Population_weights = []
    for i in Population:
        city_weight = i / Total_pop
        Population_weights.append(city_weight)
        
    return Population_weights
```

```python
Ontario_weights = population_weight(Population, Total_pop)
```

```python
Ontario_weights
```

```python
New_temp_data
```

```python
product = New_temp_data * Ontario_weights
```

```python
product
```

```python
df_product = pd.DataFrame(product.sum(axis=1))
```

```python
Total_temerature =df_product.rename(columns={0:"Total_temperature"})
```

```python
 Total_temerature
```

```python
base_temperature = 18

# Calculate cooling degree days (CDD)
Total_temerature['CDD'] = Total_temerature['Total_temperature'].apply(lambda x: max(x - base_temperature, 0))

# Calculate heating degree days (HDD)
Total_temerature['HDD'] = Total_temerature['Total_temperature'].apply(lambda x: max(base_temperature - x, 0))
```
